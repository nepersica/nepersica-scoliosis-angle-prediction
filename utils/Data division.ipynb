# 서울 삼성병원 데이터 rename


```python
dataset_path = 'External Validation/samsung-hospital/origin'
```

```python
save_path = 'External Validation/samsung-hospital/'
```

```python
import os
import cv2 as cv
import shutil
from tqdm import tqdm
```

```python
dataset = os.listdir(dataset_path)
```

```python
len(dataset)
```

```python
for data in dataset:
  data_path = os.path.join(dataset_path, data)
  image_list = os.listdir(data_path)
  if len(image_list) >= 3:
    print(data_path)
    # if os.path.isfile(os.path.join(data_path, 'image2.png')):
    #   os.remove(os.path.join(data_path, 'image2.png'))
    #   print(data_path)
```

```python
for data in tqdm(dataset):
  data_path = os.path.join(dataset_path, data)
  image_list = os.listdir(data_path)
  height_list = []
  if len(image_list) < 2:
    print(data)
    shutil.copy(os.path.join(data_path, image_list[0]), os.path.join(save_path, 'whole', f'{data}.png'))
    continue
  for image in image_list: 
    origin_image = cv.imread(os.path.join(data_path, image))
    h,w,c = origin_image.shape
    height_list.append(h)
  if height_list[0] > height_list[1]:
    shutil.copy(os.path.join(data_path, image_list[0]), os.path.join(save_path, 'whole', f'{data}.png'))
    shutil.copy(os.path.join(data_path, image_list[1]), os.path.join(save_path, 'chest', f'{data}.png'))
  elif height_list[0] < height_list[1]:
    shutil.copy(os.path.join(data_path, image_list[0]), os.path.join(save_path, 'chest', f'{data}.png'))
    shutil.copy(os.path.join(data_path, image_list[1]), os.path.join(save_path, 'whole', f'{data}.png'))
```

```python
print(len(os.listdir(os.path.join(save_path, 'whole'))))
print(len(os.listdir(os.path.join(save_path, 'chest'))))
```
